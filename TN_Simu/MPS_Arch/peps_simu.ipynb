```python
import numpy as np 
import matplotlib.pyplot as plt 
import qiskit
from qiskit import Aer
from qh_gates import *
from qh_circuit import QHCircuit
from tn_simu import TN_Architecture, PEPS_Architecture, MPS_Architecture, TN_Simulator
```

```python
def plot_state(amp):
    plt.figure()
    p = np.abs(amp) ** 2
    n = int(np.log2(len(amp)))
    label = [(r"{:0" + f"{n}" + r"b}").format(i) for i in range(len(amp))]
    plt.bar(label, p)
    plt.xticks(rotation=70)
    plt.title(f"{p.sum()}")
    plt.show()
    plt.close()
```

```python
qc = QHCircuit(2)
# qc.h(0)
# qc.h(1)
qc.cnot(1,3)
qc.cnot(3,2)
```

```python
arch = PEPS_Architecture(2, 2)
simu = TN_Simulator(arch, qc, 0)
```

```python
simu.simulate()
sv = simu.get_all_amplitudes()
plot_state(sv)
```
